##### Exploratory Data Analysis Part 1

```python
#data analysis phase
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#different color schemes
colors = plt.cm.tab10.colors
```

```python
# display all comlumns of dataset
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
```

```python
print(dataset.shape)
dataset.head()
```

### Missing Values
* dataset.columns → loops through every column name
* dataset[features].isnull() → returns True/False for each row (True = missing value)
* .sum() → adds up all the Trues → gives total count of missing values in that column
* > 1 → keeps only columns where missing count is more than 1

* Result: feature_with_nan = list of column names that have NaN values.

```python
# find out missing value pecentage of each feature/column
# use list comprehension
feature_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum() >1]
#print each feature name and percentage of missing values
for feature in feature_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean() * 100 ,4) ,'% Missing Values')
```

### Since there are many missing values so find out relationship b/w missing values and target feature i-e SalePrice

```python
# plot it
for feature in feature_with_nan:
    
    #Make a copy of the dataset so you don't accidentally modify the original dataset while experimenting.
    data=dataset.copy() 
    
    # null values refer to 1 and notnull refer to 0 in every feature of plot
    data[feature]=np.where(data[feature].isnull(), 1, 0 )

    # find meadian sale price where info is missing not mean bcz mean is average and meadian is actual approximation
    data.groupby(feature)['SalePrice'].median().plot.bar(color=['green', 'black']) 

    plt.title(feature)
    plt.show()
```

* Here relation b/w missing values and dependent variable is clearly visible so we need ro replace these nan values to a meaning full info which we done in feature selection

```python
# for above datset some of features like id is not required it is only index
print('ID of Houses {}'.format(len(dataset['Id'])))
```

### Numerical Features

```python
numerical_features=[feature for  feature in dataset.columns if dataset[feature].dtypes != 'O'] # 'o' is object so only take numercal values here 
print("No of Numerical variable " ,len(numerical_features))
dataset[numerical_features].head()
```

```python
for feature in numerical_features:
    print(feature, dataset[feature].dtype)
```

```python
# temporal variable where time or year are mentioned
year_feature=[feature for feature  in dataset.columns if 'Yr'  in feature or 'Year' in feature]
year_feature
```

```python
year_feature_data=[print(feature,dataset[feature].unique()) for  feature in year_feature]
print(year_feature_data)
    
```

```python
# find relation b/w year house is sold and saleprice
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('YrSold')
plt.ylabel('SalePrice')
```

```python
# FIND REALTION OF ALL fefatures with saleprice
for feature in year_feature:
    if feature != 'YrSold':
        data=dataset.copy()
        ##Converts the raw year into an age — e.g. YrSold - YearBuilt = "how old the house was when it sold." 
        data[feature]=data['YrSold'] - data[feature]
        plt.scatter(data[feature],data['SalePrice'] )
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

#### Numerical variables are of two types
* 1)Continuous variables >>Numbers that can take any value within a range, including decimals
* 2)Discrete Variables >>Numbers that are countable

#### Discrete Variables

```python

discrete_features=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print('Length of discrete features {}'.format(len(discrete_features)))
discrete_features
# why<25
# Discrete features (bedrooms, cars, bathrooms) usually have a small, limited set of values — like 1 to 10.
# Continuous features (price, area) usually have hundreds or thousands of unique values — almost every row is different.
```

```python
# FIND REALTION OF ALL discrete features with saleprice
for feature in discrete_features:
        data=dataset.copy()
        data.groupby(feature)['SalePrice'].median().plot.bar(color = colors)
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
    
        plt.show()
```

#### Continuous Variables

```python
continuous_features=[feature for feature in numerical_features if  feature not in discrete_features + year_feature+['Id']]
print('Length of discrete features {}'.format(len(continuous_features)))
continuous_features

```

```python
# analze continouos values by using histogram
for feature in continuous_features:
        data=dataset.copy()
        data[feature].hist(bins=25)
        plt.xlabel(feature)
        plt.ylabel('Count')
        plt.title(feature)
        plt.show()
```

##### Exploratory Data Analysis Part 2

```python
#perform logarithmic tranfromation >
# >it compresses large values and stretches small values, making skewed data more normally distributed (bell-shaped).

for feature in continuous_features:
    data=dataset.copy()
    
    if 0 in data[feature].unique(): #log(0) = -inf, which would break the transformation
        pass
    else:
        
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
        
```

```python
for feature in continuous_features:
    print(feature, dataset[feature].dtype)
```

#### Outliers
* Data points that are very different from the rest of the data — unusually high or unusually low values compared to the general pattern.


```python
for feature in continuous_features:
    data=dataset.copy()
    
    if 0 in data[feature].unique(): #log(0) = -inf, which would break the transformation
        pass
    else:
        
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
